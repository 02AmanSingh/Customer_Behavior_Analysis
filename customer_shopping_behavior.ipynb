```python
import pandas as pd

df = pd.read_csv("customer_shopping_behavior.csv")

# print(df.head())
```

```python
df.head()
```

```python
df.info()   # to get an overview of the dataset, including data types and missing values.
```

```python
# df.describe()  # to get summary statistics of the numerical columns in the dataset.

df.describe(include="all") # to get summary statistics of all columns, including categorical ones.
```

```python
df.isnull().sum() # to check for missing values in each column of the dataset.
```

```python
df["review_rating"] = df.groupby("category")["review_rating"].transform(lambda x: x.fillna(x.median())) # to fill missing values in the "Review Rating" column with the median rating for each category.
```

```python
df.isnull().sum()  # to check for missing values in each column of the dataset.
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns = {"purchase_amount_(usd)" : "purchase_amount"})
```

```python
df.columns
```

```python
#create new column age_group based on age column
labels =["Young-Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"] = pd.qcut(df["age"], q=4, labels=labels)
df[["age", "age_group"]].head(10)
```

```python
#create a new column purchase_frequency_days

frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Annually": 365,
    "Every 3 Months": 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days", "frequency_of_purchases"]].head(10)
```
